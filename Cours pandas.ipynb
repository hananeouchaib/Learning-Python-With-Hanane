```
# remplacer les virgules par rien dans la colonne Salary
table['Salary'] = table['Salary'].replace(',','', regex=True)
```

```
# convertir le type object de la colonne Salary en int
table['Salary'] = pd.to_numeric(table['Salary'])
```

# **loc et iloc**

```
# affiche toutes les valeurs de la 4√®me ligne
table.iloc[4]
```

```
# affiche toutes les valeurs des lignes entre 10 et 19
table.iloc[10:20]
```

```
# affiche les valeurs de la colonne 0 des lignes 10 jusqu'√† 19
table.iloc[10:20,0]
```

```
# affiche les valeurs des deux colonnes salary et manager pour les lignes 10 jusqu'√† 20 inclu
table.loc[10:20,['Salary','Manager']]
```
